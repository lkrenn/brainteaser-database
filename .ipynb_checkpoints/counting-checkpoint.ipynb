# Poker Hand Probabilities
What is the probability of being dealt a certain combination in a 5-card poker hand? In a standard deck, there are 52 cards with 4 suits, 13 "numbers" thus there are:
$$
52 \choose 5
$$
Total combinations of 5-card hands.

## Royal Flush
There are only 4 suits and one way per suit to get a royal flush. Thus, the answer is $4$

## Straight Flush
There are 4 suits in a deck and 10 ways (don't forget an ace counts as 1!) to get 5 numbers in a row. However, we have to be sure to not include royal flushes, thus the answer is:
$$
4\text{ suits }\times 10\text{ possibilities } - 4\text{ straight flushes } = 36
$$

## 4 of a Kind
$$
\begin{align}
{13 \choose 1} &\text{ ways to pick the 4-of-kind number} \\
{4 \choose 4} &\text{ ways to pick the suits of the 4-of-kind}\\
{12 \choose 1} &\text{ ways to pick the remaining number} \\
{4 \choose 1} &\text{ ways to pick the suit of the remaining number}
\end{align}
$$
Therefore we get:
$$
\frac{{13 \choose 1}{4 \choose 4}{12 \choose 1}{4 \choose 1}}{52 \choose 5} \approx 0.024\%
$$

## Full House
$$
\begin{align}
{13 \choose 1} &\text{ ways to pick the 3-of-kind number} \\
{4 \choose 3} &\text{ ways to pick the suits of the 3-of-kind}\\
{12 \choose 1} &\text{ ways to pick the pair number} \\
{4 \choose 2} &\text{ ways to pick the suit of the pair}
\end{align}
$$
Therefore we get:
$$
\frac{{13 \choose 1}{4 \choose 3}{12 \choose 1}{4 \choose 2}}{52 \choose 5} \approx 0.144\%
$$

## Flush
$$
\begin{align}
{4 \choose 1} &\text{ ways to pick the suit of the flush} \\
{13 \choose 5} &\text{ ways to pick numbers}\\
36 &\text{ ways to make a straight flush (don't count!)} \\
4 &\text{ ways to make a royal flush (don't count!)}
\end{align}
$$
Therefore we get:
$$
\frac{{4 \choose 1}{13 \choose 5}-40}{52 \choose 5} \approx 0.1965\%
$$

## Straight
$$
\begin{align}
10 &\text{ ways to pick 5 consecutive numbers} \\
4^5 &\text{ ways to pick the suits of the numbers}\\
36 &\text{ ways to make a straight flush (don't count!)} \\
4 &\text{ ways to make a royal flush (don't count!)}
\end{align}
$$
Therefore we get:
$$
\frac{10\cdot4^5-40}{52 \choose 5} \approx 0.3925\%
$$

## 3 of a Kind
$$
\begin{align}
{13 \choose 1} &\text{ ways to pick the 3-of-kind number} \\
{4 \choose 3} &\text{ ways to pick the suits of the 3-of-kind}\\
{12 \choose 2} &\text{ ways to pick the remaining 2 numbers (no pairs!)} \\
{4 \choose 1} &\text{ ways to pick the suit of the first remaining card} \\ 
{4 \choose 1} &\text{ ways to pick the suit of the second remaining card} \\ 
\end{align}
$$
Therefore we get:
$$
\frac{{13 \choose 1}{4 \choose 3}{12 \choose 2}{4 \choose 1}{4 \choose 1}}{52 \choose 5} \approx 2.1128\%
$$

## 2 Pairs
$$
\begin{align}
{13 \choose 2} &\text{ ways to pick the pair numbers} \\
{4 \choose 2} &\text{ ways to pick the suits of the first pair}\\
{4 \choose 2} &\text{ ways to pick the suits of the second pair}\\
{11 \choose 1} &\text{ ways to pick the number of the remaining card}\\
{4 \choose 1} &\text{ ways to pick the suit of the remaining card} \\ 
\end{align}
$$
Therefore we get:
$$
\frac{{13 \choose 1}{4 \choose 3}{12 \choose 2}{4 \choose 1}{4 \choose 1}}{52 \choose 5} \approx 4.7539\%
$$

## 1 Pair
$$
\begin{align}
{13 \choose 1} &\text{ ways to pick the pair number} \\
{4 \choose 2} &\text{ ways to pick the suits of the pair}\\
{12 \choose 3} &\text{ ways to pick the numbers of the remaining card}\\
{4 \choose 1} &\text{ ways to pick the suit of the first remaining card}\\
{4 \choose 1} &\text{ ways to pick the suit of the second remaining card} \\ 
{4 \choose 1} &\text{ ways to pick the suit of the third remaining card} \\ 
\end{align}
$$
Therefore we get:
$$
\frac{{13 \choose 1}{4 \choose 2}{12 \choose 3}{4 \choose 1}{4 \choose 1}{4 \choose 1}}{52 \choose 5} \approx 42.2569\%
$$
